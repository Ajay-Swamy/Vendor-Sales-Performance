## Dataset Overview

- The **purchases** table contains actual purchase data, including:
  - Purchase date
  - Products (brands) purchased by vendors
  - Amount paid (in dollars)
  - Quantity purchased

- The **purchase_prices** column is derived from purchase_prices table, which provides product wise :
  - Actual price
  - Purchase price
  - A unique combination of **Vendor** and **Brand**

- The **vendor_invoice** table is an aggregated version of the purchases data and includes:
  - Total quantity purchased
  - Total dollar amount
  - Freight cost
  - Unique records based on **Vendor** and **PO Number**

- The **sales** table contains actual sales transactions, including:
  - Brands purchased
  - Quantity sold
  - Selling price
  - Revenue generated

---

## Objective

Since the required analysis data is distributed across multiple tables, we need to create a consolidated summary table containing:

- Purchase transactions made by vendors
- Sales transaction data
- Freight costs for each vendor
- Actual product prices from vendors

## Creating a final Summary Table

```python
vendor_sales_summary=pd.read_sql_query("""with freightSummary as(
select VendorNumber,
        sum(Freight) as FreightCost
from vendor_invoice
group by VendorNumber
),

PurchaseSummary as(
    select
        p.VendorNumber,
        p.VendorName,
        p.Brand,
        p.Description,
        p.PurchasePrice,
        pp.Price as ActualPrice,
        pp.Volume,
        sum(p.Quantity) as TotalPurchaseQuantity,
        sum(p.Dollars) as TotalPurchaseDollars
    from purchases p
    join purchase_prices pp
    on p.Brand=pp.Brand
    where p.PurchasePrice >0
    group by p.VendorNumber,
        p.VendorName,
        p.Brand,
        p.Description,
        p.PurchasePrice,
        pp.Price,
        pp.Volume
),

SalesSummary as(
    select 
        VendorNo,
        Brand,
        sum(SalesQuantity) as TotalSalesQuantity,
        sum(SalesDollars) as TotalSalesDollars,
        sum(SalesPrice) as TotalSalesPrice,
        sum(ExciseTax) as TotalExciseTax
    from sales
    group by VendorNo,Brand
)
    SELECT
    ps.VendorNumber,
    ps.VendorName,
    ps.Brand,
    ps.Description,
    ps.PurchasePrice,
    ps.ActualPrice,
    ps.Volume,
    ps.TotalPurchaseQuantity,
    ps.TotalPurchaseDollars,
    ss.TotalSalesQuantity,
    ss.TotalSalesDollars,
    ss.TotalSalesPrice,
    ss.TotalExciseTax,
    fs.FreightCost
FROM PurchaseSummary ps
LEFT JOIN SalesSummary ss
    ON ps.VendorNumber = ss.VendorNo
    AND ps.Brand = ss.Brand
LEFT JOIN FreightSummary fs
    ON ps.VendorNumber = fs.VendorNumber
ORDER BY ps.TotalPurchaseDollars DESC""",con)
```

## Why Create the `vendor_sales_summary` Table?

The `vendor_sales_summary` table is designed to improve analytical performance and simplify reporting by consolidating vendor purchase, sales, pricing, and freight information into a single dataset.

### Benefits

- Reduces query execution time by avoiding repeated joins and aggregations.
- Provides a centralized dataset for vendor performance analysis.
- Supports faster dashboard refreshes in Power BI.
- Simplifies KPI calculations related to sales, purchases, profit margins, and freight costs.
- Improves scalability when working with large transactional datasets.
- Serves as a reliable source for reporting and business decision-making.

```python
# removed white spaces at the end of Vendor Name
vendor_sales_summary['VendorName']=vendor_sales_summary['VendorName'].str.strip()
```

#### Adding extra columns to the vendor_sales_summary data frame
- GrossProfit
- Profit Margin
- Stock Turnover
- sales to purchase ratio

```python
vendor_sales_summary['GrossProfit']=vendor_sales_summary['TotalSalesDollars']-vendor_sales_summary['TotalPurchaseDollars']
vendor_sales_summary['ProfitMargin']=(vendor_sales_summary['GrossProfit']/vendor_sales_summary['TotalSalesDollars'])*100
vendor_sales_summary['StockTurnOver']=vendor_sales_summary['TotalSalesQuantity']/vendor_sales_summary['TotalPurchaseQuantity']
vendor_sales_summary['SalesToPurchaseRatio']=vendor_sales_summary['TotalSalesDollars']/vendor_sales_summary['TotalPurchaseDollars']
```

#### making the vendor_sales_summary as Table and ingesting it into DataBase

```python
cursor=con.cursor()
```

```python
cursor.execute("""
CREATE TABLE vendor_sales_summary (
    VendorNumber INT,
    VendorName VARCHAR(100),
    Brand INT,
    Description VARCHAR(100),
    PurchasePrice DECIMAL(10,2),
    ActualPrice DECIMAL(10,2),
    Volume INT,
    TotalPurchaseQuantity INT,
    TotalPurchaseDollars DECIMAL(15,2),
    TotalSalesQuantity INT,
    TotalSalesDollars DECIMAL(15,2),
    TotalSalesPrice DECIMAL(15,2),
    TotalExciseTax DECIMAL(15,2),
    FreightCost DECIMAL(15,2),
    GrossProfit DECIMAL(15,2),
    ProfitMargin DECIMAL(15,2),
    StockTurnover DECIMAL(15,2),
    SalesToPurchaseRatio DECIMAL(15,2),
    PRIMARY KEY (VendorNumber, Brand)
);
""")
```
