```python
import pandas as pd
import os
from sqlalchemy import create_engine
import logging
import time
```

```python
engine=create_engine('sqlite:///databases/inventory.db');
```

```python
'''this function will ingest the data frames into database'''
def ingest_into_db(df,table_name,engine):
    df.to_sql(table_name,con=engine,if_exists='replace',index=False);
```

```python
'''this will load csv's as dataframe and ingest into database'''
def load_raw_data():
    for file in os.listdir('Data'):
        if file.ends_with('.csv'):
            df=pd.read_csv('Data/'+file);
            print(f'ingesting file{file} in database....');
            print(df.shape);
            ingest_into_db(df,file[:-4],engine);
```

```python
if __name__=='__main__':
    load_raw_data()
```
